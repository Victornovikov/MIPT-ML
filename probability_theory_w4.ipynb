## Random variables and random verctors
$(\Omega, \mathcal{F}, P)$ -- probability space, 
$\mathcal{E}$ - σ-algebra on E.

- If f: Ω -> E is $(\mathcal{F}|\mathcal{E})$-measurable, then this function is called a random element. 




Now we can consider some special cases:

E = $\mathbb{R}$, $\mathcal{E} = \mathcal{B}(\mathbb{R}) \Rightarrow$ f - a random variable,

E = $\mathbb{R}^n$, $\mathcal{E} = \mathcal{B}(\mathbb{R^n}) \Rightarrow$ f - a random vector.



## Why do we need measurability?
When we talk about descrete cases we don't really need it. 

For example, we can consider a random variable that is equal to a number that dice shows. We flip a dice, we look at a number, and we say that the number is our random variable that takes values {1,2,3,4,5,6} and we can forget about measuarability, but only because it holds.

When we start working with events that has much more values it becomes difficult.

For example lets take f -- temperature tomorrow at 5pm. 

f is a real number. We want to compute probabilites that are related to some intuitive sets. For example, we may want to compute probability that temperature is between 20 and 21 degrees.

What does it mean?
Normally have probability space and some events. But in this case we're dealing with something different -- have a probability of some inequality.
 
In this case it is a probability of all elementary events s.t. our temperature (random variable on these elementary events) belongs to our set. 



$f$ - temperature

$ P(20 \leq f \leq 21) = 
 P(\underbrace{\{ \omega: 20 \leq f(\omega) \leq 21 \}}_{\subset \Omega}) $

This is a set in Ω, but why can we measure its probability?
We can measure probability of a set if it belongs to F.
$ \{ \omega: 20 \leq \underbrace{f(\omega)}_{= f^{-1}([20,21])} \leq 21 \} \in \mathcal{F}$

Interval [20,21] is a borel set. By our definitino it is TRUE -- for every $f^{-1}$ of this set belongs to f.
This is why we need this notion of measurability.

##Probability distribution
https://www.youtube.com/watch?v=u24fXSHTR10


We discussed why we needed the notion of measurability on the example of measuring temperature. We needed to find a probability that temperature belongs to some set. What it means is that we have some probability distribution that is induced by our random variable or a random vector. 

For example, in the case of measuring temperature we looked at a set $\omega$ of all elementary events, s.t. $\xi(\omega)$ belongs to some interval. In other words, we computed a probability that random variable belongs to an interval [20, 21], or we can say that this probability is a probability on $\mathbb{R}$ -- *we can forget about our initial probability* and say that now we have a new probability distirbuiton that is induced by our random variable.

That means that the following is true:
If we consider a function  $P_\xi$, s.t. for every Borel set the value of this function equals the probability that $\xi$ belongs to B, this is also the probability of $\xi^{-1}(B)$ (the probability of the set of all $\omega$, s.t. $\xi(\omega)$ belongs to B.

If we define the function in this way, then this function is *a probability*.

### Theorem
The function 

$P_{\xi}: \mathcal{B}(\mathbb{R}^n) \rightarrow [0,1], P_{\xi}(B) = P(\xi \in B)\\
\text{is a probability on } (\mathbb{R}, \mathcal{B}(\mathbb{R}^n)).$

To prove that we need to check properties of probability.

image.png

- Since it ($P_{\xi}$) is a probability distribution we can call it "distribution of $\xi$"

- The distribution function $F_ξ$ of $P_ξ$ is called distribution function of ξ. 
- If $P_ξ$ is absolutely continuous, then its density $p_ξ$ is called density of ξ, ξ is called absolutely continuous
-If $P_ξ$ is discrete, then ξ is called discrete as well.


### Notations 
P({ω : ξ(ω) ∈ B}) =: P(ξ ∈ B). 

("a probability of such a set $\omega$ that $\xi(\omega) \in B$")

In particular, 

1. P({ω : ξ(ω) < x}) =: P(ξ < x)

2. P({ω : ξ(ω) = x}) =: P(ξ = x)

3. Fξ(x) = P({ω : ξ(ω) ≤ x}) =: P(ξ ≤ x).
